matplotlib.pyplot.bar(x, height, width=0.8, bottom=None, , align='center', data=None, *kwargs)

参数说明： x：包含所有柱子的下标的列表

height：包含所有柱子的高度值的列表

width：0~1，默认0.8

botton：条形的起始位置，也是y轴的起始坐标

align：柱子对齐方式，有两个可选值：center和edge边缘，默认值是center

其他可选参数有： color：条形的颜色 “r","b","g","#123465"，默认“b"蓝色

edgecolor：每根柱子边框的颜色，同上

linewidth：每根柱子的边框宽度，默认是没有边框。

tick_label：下标的标签，默认是没有内容。

xerr每根柱子顶部在横轴方向的线段。如果指定一个固定值，所有柱子的线段将一直长；如果指定一个带有不同长度值的列表，那么柱子顶部的线段将呈现不同长度。

yerr每根柱子顶端在纵轴方向的线段。如果指定一个固定值，所有柱子的线段将一直长；如果指定一个带有不同长度值的列表，那么柱子顶部的线段将呈现不同长度。

alpha：透明度 越小颜色越淡

```python
#matplotlib 官方提供的例子

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from collections import namedtuple

n_groups = 5

means_men = (20, 35, 30, 35, 27)
std_men = (2, 3, 4, 1, 2)

means_women = (25, 32, 34, 20, 25)
std_women = (3, 5, 2, 3, 3)

fig, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.35

error_config = {'ecolor': '0.3'}

rects1 = ax.bar(index, means_men, bar_width,
                alpha=0.4, color='b',
                yerr=std_men, error_kw=error_config,
                label='Men')

rects2 = ax.bar(index + bar_width, means_women, bar_width,
                alpha=0.4, color='r',
                yerr=std_women, error_kw=error_config,
                label='Women')   #alpha透明度

ax.set_xlabel('Group')
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(('A', 'B', 'C', 'D', 'E'))
ax.legend()

fig.tight_layout()
plt.show()
```

如何在一张图里面，同时反映经济总量的同时，再加上经济增速的变化？

堆叠条形图 + 折线图

用上一篇文章提到的，twinx()方法，绘制完bar()，后共用一个x轴，再另外绘制一个plot()折线图即可。

```python
y_2019_GDP= [12406.8,13908.57,11879.22,6786.87,9143.64]
y_2018_GDP= [11406.8,12908.57,9089.77,5986.87,7043.64]
data=[0.26,0.38,0.28,0.95,0.80]
wid=0.35
area=['北京市','上海市','深圳市','西安市','重庆市']

fig=plt.figure(figsize=(12,7))

ax1=fig.add_subplot(111)
#画图
plt.bar(np.arange(5), #x轴数据
        y_2019_GDP,  #y轴数据
        color='blue', # 颜色
        width=wid,
        label = 'y_2019_GDP',
        alpha = 0.8)   # 透明度  越小颜色越淡
plt.bar(np.arange(5), #x轴数据
        y_2018_GDP,  #y轴数据
        align ='center', #设置条形的中心位置在中间，edge为边缘
        color='red', # 颜色
        width=wid,
        label = 'y_2018_GDP',
        bottom=y_2019_GDP,
        alpha = 0.8)   # 透明度  越小颜色越淡

ax2 = ax1.twinx()
z=np.array(list(data))
x=np.arange(len(area))
ax2.plot(x,z,c='k',marker='*',markersize =12,markerfacecolor='yellow',markeredgecolor='red')
#打印折线图的数据
for a,b in zip(x,z):
    plt.text(a, b, '%s' % b, ha='center', va= 'bottom',fontsize=14)    
# 坐标轴名称
ax1.set_xlabel("区域",fontdict={'fontsize':16,'color':'black'})
ax1.set_ylabel("GDP",fontdict={'fontsize':16,'color':'black'})
ax2.set_ylabel("增长百分比",fontdict={'fontsize':16,'color':'black'})
# # 设置标题
plt.title("五大城市GDP总量以及经济增速统计",fontdict={'fontsize':20,'color':'black'})
#添加图例
ax1.legend(loc='best')   
# 添加刻度标签
plt.xticks(np.arange(5),area)
plt.show()
#保存成图片
plt.savefig("result.png",format='png')
```

```python
index = np.arange(8)

fig = plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(111) 

ax1.bar(index, result['人数'],color='b',alpha=0.3,label='人数')
ax1.set_ylim([0, 8000])  #设置y轴取值范围

ax2 = ax1.twinx() # this is the important function  
ax2.plot(index,result['人数占比'],'or-',label='人数占比')
ax2.set_ylim([0, 0.35])  #设置y轴取值范围
for i,(x,y) in enumerate(zip(index,result['人数占比'])):  
    plt.text(x+0.05,y+0.001,np.round(result['人数占比'][i],2),color='black',fontsize=12)  #将数值显示在图形上

plt.xticks(index,result['客户类型'])  #设置中文标签
# 坐标轴名称
#ax1.set_xlabel("客户",fontdict={'fontsize':16,'color':'black'})
ax1.set_ylabel('人数',fontdict={'fontsize':16,'color':'black'})
ax2.set_ylabel("人数占比",fontdict={'fontsize':16,'color':'black'})
#添加图例
ax1.legend(loc=2)
ax2.legend(loc=1)
plt.title('各类型客户人数分布及占比',fontdict={'fontsize':18,'color':'black'})
```
